```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
photo_r_min = 1000
photo_r_max = 100000
V_in = 3.3
```

```python
def voltage_divide(V_in, R1, R2):
    return (R2 / (R1 + R2)) * V_in
```

```python
Rs = np.arange(10000)
```

```python
V_out_min = voltage_divide(V_in, photo_r_min, Rs)
V_out_max = voltage_divide(V_in, photo_r_max, Rs)

fig = plt.figure(figsize=(11, 8))
ax = fig.add_subplot(111)
ax.plot(Rs, V_out_min, label='Minimum Measured Voltage')
ax.plot(Rs, V_out_max, label='Maximum Measured Voltage')
ax.legend()
ax.grid()
ax.set_xlabel('Resistor2 in Ohms')
ax.set_ylabel('V Measured')
```
